### The method of extraction: 
Here, the column name is extracted by **finding the column_name that has the maximum number of words in each row**. Since, the review msg column would most probably contain lot of words compared to other column.<br> 
It is checked for each row, where each row would give a column_name that has the most no.of words. And the most occuring column_name from half of the dataset is <b><u>chosen<u><b>

```python
import pandas as pd
import numpy as np
import statistics
```

```python
#enter name of the csv file
x=input("enter file name(csv):")
file_name='/Users/shwethailango/Documents/Work/sentalyze/'+x+'.csv'

#read file into a dataframe
df=pd.read_csv(file_name)

#list to append the column name that has the most number of words in each row
a=[]

#iterating the dataframe
for i in range(0,int(len(df)/2)):
    count = df.iloc[i].str.split().str.len() #'count' stores the number of words for each column in a row.
    a.append(count[count==max(count)].index[0]) #column name with maximum 'count' is stored in list'a'.
print("Preferred column name would be ",statistics.mode(a))#the elt with maximum occurence in 'a' (column_name) is taken as output.
```

#### Thus the preferred column name is **reviews.text** and this can be further sent to user for confirmation
